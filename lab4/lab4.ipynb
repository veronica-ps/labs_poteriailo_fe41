```python
initial_amplitude = 1.0
initial_frequency = 0.5
initial_phase = 0.0
initial_noise_mean = 0.0
initial_noise_covariance = 0.1
initial_cutoff = 2.0
initial_show_noise = True

# Створюємо вісь часу
t = np.linspace(0, 10, 1000)
fs = 1000 / 10.0  # Частота дискретизації

base_noise = np.random.randn(len(t))

def harmonic_with_noise(amplitude, frequency, phase, noise_mean, noise_covariance, show_noise):
    clean_signal = amplitude * np.sin(2 * np.pi * frequency * t + phase)

    if show_noise:
        actual_noise = base_noise * np.sqrt(noise_covariance) + noise_mean
    else:
        actual_noise = np.zeros_like(t)

    noisy_signal = clean_signal + actual_noise
    return clean_signal, noisy_signal

def apply_filter(data, cutoff, fs, order=4):
    nyquist = 0.5 * fs
    normal_cutoff = cutoff / nyquist
    if normal_cutoff >= 1.0: normal_cutoff = 0.99
    if normal_cutoff <= 0.0: normal_cutoff = 0.01

    b, a = butter(order, normal_cutoff, btype='low', analog=False)
    return filtfilt(b, a, data)

fig, ax = plt.subplots(figsize=(12, 7))
plt.subplots_adjust(left=0.1, bottom=0.45, right=0.8) # Місце для UI
ax.set_title("Інтерактивна гармоніка з шумом та фільтрацією")
ax.set_xlabel("Час (t)")
ax.set_ylabel("Амплітуда")

clean_sig, noisy_sig = harmonic_with_noise(
    initial_amplitude, initial_frequency, initial_phase,
    initial_noise_mean, initial_noise_covariance, initial_show_noise
)
filtered_sig = apply_filter(noisy_sig, initial_cutoff, fs)

# Відображення ліній на графіку
line_noisy, = ax.plot(t, noisy_sig, lw=1.5, color='orange', label='Зашумлена гармоніка')
line_clean, = ax.plot(t, clean_sig, lw=2, color='blue', linestyle='--', label='Чиста гармоніка')
line_filtered, = ax.plot(t, filtered_sig, lw=2.5, color='purple', label='Відфільтрована гармоніка')
ax.legend(loc='upper right')

# слайдери
axcolor = 'lightcyan'
ax_amp = plt.axes([0.15, 0.35, 0.60, 0.03], facecolor=axcolor)
ax_freq = plt.axes([0.15, 0.30, 0.60, 0.03], facecolor=axcolor)
ax_phase = plt.axes([0.15, 0.25, 0.60, 0.03], facecolor=axcolor)
ax_mean = plt.axes([0.15, 0.18, 0.60, 0.03], facecolor='mistyrose')
ax_cov = plt.axes([0.15, 0.13, 0.60, 0.03], facecolor='mistyrose')
ax_cutoff = plt.axes([0.15, 0.06, 0.60, 0.03], facecolor='thistle')

samp = Slider(ax_amp, 'Amplitude', 0.1, 5.0, valinit=initial_amplitude)
sfreq = Slider(ax_freq, 'Frequency', 0.1, 5.0, valinit=initial_frequency)
sphase = Slider(ax_phase, 'Phase', 0.0, 2*np.pi, valinit=initial_phase)
smean = Slider(ax_mean, 'Noise Mean', -1.0, 1.0, valinit=initial_noise_mean)
scov = Slider(ax_cov, 'Noise Covariance', 0.0, 2.0, valinit=initial_noise_covariance)
scutoff = Slider(ax_cutoff, 'Filter Cutoff', 0.1, 15.0, valinit=initial_cutoff)

# чекбоккси
ax_check = plt.axes([0.82, 0.25, 0.15, 0.15])
check = CheckButtons(ax_check, ['Show Noise', 'Show Filtered'], [True, True])

# функція оновлення
def update(val):
    amp = samp.val
    freq = sfreq.val
    phase = sphase.val
    mean = smean.val
    cov = scov.val
    cutoff = scutoff.val

    show_n = check.get_status()[0]
    show_f = check.get_status()[1]

    c_sig, n_sig = harmonic_with_noise(amp, freq, phase, mean, cov, show_noise=show_n)

    # Оновлення даних на графіку
    line_clean.set_ydata(c_sig)

    if show_n:
        line_noisy.set_ydata(n_sig)
        line_noisy.set_visible(True)
    else:
        line_noisy.set_visible(False)

    if show_f:
        # Фільтруємо зашумлений сигнал
        f_sig = apply_filter(n_sig, cutoff, fs)
        line_filtered.set_ydata(f_sig)
        line_filtered.set_visible(True)
    else:
        line_filtered.set_visible(False)

    fig.canvas.draw_idle()

# Прив'язуємо функцію оновлення
samp.on_changed(update)
sfreq.on_changed(update)
sphase.on_changed(update)
smean.on_changed(update)
scov.on_changed(update)
scutoff.on_changed(update)
check.on_clicked(update)

# кнопка reset
ax_reset = plt.axes([0.02, 0.06, 0.08, 0.05])
btn_reset = Button(ax_reset, 'Reset', color='lightgrey', hovercolor='0.9')

def reset_sliders(event):
    samp.reset()
    sfreq.reset()
    sphase.reset()
    smean.reset()
    scov.reset()
    scutoff.reset()
    # Відновлюємо чекбокси
    if not check.get_status()[0]: check.set_active(0)
    if not check.get_status()[1]: check.set_active(1)

btn_reset.on_clicked(reset_sliders)

plt.show()
```
